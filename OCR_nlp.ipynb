# **P.S: This input image is of single word**

**1. READ IMAGES**

```python
import matplotlib.pyplot as plt
import cv2
import easyocr
import numpy as np
```

```python
Image_path=("/content/surf.jpeg")
```

```python
reader=easyocr.Reader(["en"], gpu=False)
result=reader.readtext(Image_path)
result
```

**Accuracy:** 
SURF -  0.98

**2. DRAW RESULT**

```python
top_left=tuple(result[0][0][0])
bottom_right=tuple(result[0][0][2])
text=result[0][1]
font=cv2.FONT_HERSHEY_SIMPLEX
```

```python
img=cv2.imread(Image_path)
img=cv2.rectangle(img, top_left,bottom_right,(0,255,0),5)
img=cv2.putText(img,text,top_left,font,0.75,(255,0,255),2,cv2.LINE_AA)
plt.imshow(img)
plt.show()
```

# **P.S: This input image is for multiple words**

**3. HANDLING MULTIPLE LINES**

```python
Image=("/content/sheprali3.jpg")
```

```python
reader1=easyocr.Reader(["en"], gpu=False)
result1=reader1.readtext(Image)
result1
```

**ACCURACY**:
*  'CHAITALI'- 0.99 
*   '01'     - 0.93

*  'SHeeTAL'- 0.77
*   '02'   - 0.99

*  'PRAJAKTA' - 0.74
*   '03'      - 0.84







```python
imgg = cv2.imread(Image)
for detection in result1: 
    top_left = tuple([int(val) for val in detection[0][0]])
    bottom_right = tuple([int(val) for val in detection[0][2]])
    text = detection[1]
    font=cv2.FONT_HERSHEY_SIMPLEX
    imgg = cv2.rectangle(imgg,top_left,bottom_right,(0,0,0),5)
    imgg = cv2.putText(imgg,text,top_left,font,3,(255,0,0),5,cv2.LINE_AA)
    
#plt.figure(figsize=(20,20))    
plt.imshow(imgg)
plt.show()
```
